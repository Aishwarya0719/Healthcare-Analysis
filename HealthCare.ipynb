```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(df['patient_id'],bins=10)
plt.show()
```

```python
#Billing Amount Distribution
sns.histplot(df["billing_amount"],bins=30,kde=True)
plt.title("Distribution of Billing Amount")
plt.show()
```

```python
#Bed  OCcupancy vs Billimng Amount
sns.boxplot(x=df["bed_occupancy"], y=df["billing_amount"])
plt.title("Bed Occupancy vs Billing Amount")
plt.show()
```

```python
#Most Common Diagnoses
sns.countplot(y=df["doctors_diagnosis"], order=df["doctors_diagnosis"].value_counts().index)
plt.title("Most Common Diagnoses")
plt.show()
```

```python
#health Insurance Distribution
sns.countplot(x=df["health_insurance"])
plt.title("patients with health insurance")
plt.show()
```

```python
#patient Admissions over time
df["admit_date"].value_counts().sort_index().plot(figsize=(12,5))
plt.title("Patient Admissions Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Admissions")
plt.show()
```

```python
#Feedback Analysis
sns.countplot(x=df["feedback"])
plt.title("Patient feedaback Distribution")
plt.show()
```

```python
#Billing Amount By Health Insurance status
sns.boxplot(x=df["health_insurance"], y=df["billing_amount"])
plt.title("Billing Amount by health Insurance status")
plt.show()
```

```python
#Most Frequent Medical tests
sns.countplot(y=df["tests"], order=df["tests"].value_counts().index)
plt.title("Most Commn Medical Test")
plt.show()
```
